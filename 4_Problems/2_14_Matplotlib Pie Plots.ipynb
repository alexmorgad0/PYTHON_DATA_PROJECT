```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## ðŸŸ© Health Insurance Pie Chart (2.14.1)

```python
df_DA = df[df['job_title_short'] == 'Data Analyst']

df_counts = df_DA['job_health_insurance'].value_counts()

plt.pie(df_counts, labels=['No', 'Yes'], autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Data Analyst Jobs Mentioning Health Insurance')
plt.show()
```

---

## ðŸŸ¨ Job Schedule Distribution (2.14.2)

```python
df_ft_pt = df[(df['job_schedule_type'] == 'Full-time') | (df['job_schedule_type'] == 'Part-time')].copy()
job_schedule_counts = df_ft_pt['job_schedule_type'].value_counts()
plt.pie(job_schedule_counts, labels=job_schedule_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Job Schedule Distribution Full-Time vs. Part-Time')
plt.show()
```

---

## ðŸŸ¥ Degree Mentioned Pie Chart (2.14.3)

```python
df_US = df[(df['job_country'] == 'United States')].copy()
job_degree_counts = df_US['job_no_degree_mention'].value_counts()
labels = ['Degree Mentioned' if not val else 'No Degree Mentioned' for val in job_degree_counts.index]

plt.pie(job_degree_counts, explode = (0, 0.1), labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Jobs Mentioning Degrees for US Job Postings')
plt.show()
```
